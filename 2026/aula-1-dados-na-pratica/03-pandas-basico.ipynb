## 1. DataFrame: a planilha em código

Um **DataFrame** é uma tabela em memória. Tem linhas e colunas, cabeçalho, tipos.

Vamos carregar nossos clientes:

```python
import pandas as pd

df = pd.read_csv("data/clientes_pequeno.csv")
df.head()
```

## 2. Os 4 comandos que resolvem 80%

```python
# Quantas linhas? Quantas colunas? Que tipos?
df.info()
```

```python
# Estatísticas resumidas das colunas numéricas
df.describe()
```

```python
# Quantos registros por categoria?
df["categoria"].value_counts()
```

```python
# Os 10 maiores valores de compra
df.nlargest(10, "valor_compra")
```

## 3. Filtros — "WHERE" do Pandas

Você usa colchetes com uma condição booleana.

```python
# Clientes de São Paulo
df_sp = df[df["estado"] == "SP"]
df_sp.head()
```

```python
# Clientes que gastaram mais de R$ 1.000
df_grandes = df[df["valor_compra"] > 1000]
print(f"{len(df_grandes)} clientes gastaram mais de R$ 1.000")
df_grandes.head()
```

```python
# Combinando filtros (AND com &, OR com |)
df_sp_caros = df[(df["estado"] == "SP") & (df["valor_compra"] > 500)]
df_sp_caros
```

## 4. GroupBy — agregando por categoria

A operação mais útil do Pandas. "Para cada X, calcule Y.

```python
# Total gasto por estado
df.groupby("estado")["valor_compra"].sum().sort_values(ascending=False)
```

```python
# Quantas compras e valor médio por categoria
df.groupby("categoria").agg(
    quantidade=("id", "count"),
    valor_medio=("valor_compra", "mean"),
).round(2)
```

## 5. Salvar resultado

Análise pronta? Salve em Parquet pra usar depois (ou compartilhar).

```python
resumo = df.groupby("estado")["valor_compra"].sum().reset_index()
resumo.to_parquet("data/resumo_por_estado.parquet")
print("Salvo em data/resumo_por_estado.parquet")
```
