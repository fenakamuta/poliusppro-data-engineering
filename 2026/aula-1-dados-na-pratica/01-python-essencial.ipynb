## 1. Variáveis

Variável é uma "caixa" com nome onde você guarda um valor. Em Python, você cria com `=`.

```python
# Texto (string)
nome = "Ana"

# Número
idade = 32

# Mostrando o valor
print(nome, "tem", idade, "anos")
```

## 2. Listas

Uma lista é uma coleção ordenada. Use `[ ]`.

```python
clientes = ["Ana", "João", "Maria"]

# Acessando por posição (começa em 0)
print("Primeiro:", clientes[0])
print("Total:", len(clientes), "clientes")

# Iterando
for c in clientes:
    print("- ", c)
```

### Dicionários

Coleção com chave → valor. Use `{ }`.

```python
cliente = {
    "nome": "Ana",
    "cidade": "São Paulo",
    "compra": 150.00,
}

print(cliente["nome"], "comprou", cliente["compra"])
```
