Objetivo: entender os **três formatos mais comuns** de dado e quando usar cada um.

Você vai ler o mesmo dado nos três formatos e comparar.

---

## Dados de exemplo

Na pasta `data/` deste notebook há três arquivos com **o mesmo conteúdo** (100 clientes):
- `clientes_pequeno.csv`
- `clientes_pequeno.json`
- `clientes_pequeno.parquet`

## 1. CSV (Comma-Separated Values)

O "feijão com arroz". Arquivo de texto onde cada linha é um registro e as colunas são separadas por vírgula.

```python
import pandas as pd

df_csv = pd.read_csv("data/clientes_pequeno.csv")
df_csv.head()
```

**O bom:** qualquer ferramenta abre (Excel, Notepad, Python).

**O ruim:** sem tipo de dado (tudo é texto), arquivos enormes, sem compressão.

## 2. JSON (JavaScript Object Notation)

Formato textual em estrutura de chave → valor. Padrão de APIs modernas.

```python
df_json = pd.read_json("data/clientes_pequeno.json")
df_json.head()
```

**O bom:** suporta estrutura aninhada (listas dentro de objetos).

**O ruim:** ainda é texto e pesado para dados tabulares grandes.

## 3. Parquet — o formato dos profissionais

Formato **colunar e comprimido**. Otimizado para leitura analítica.

```python
df_parquet = pd.read_parquet("data/clientes_pequeno.parquet")
df_parquet.head()
```

**O bom:** rápido, leve, preserva tipos (datas são datas, números são números).

**O ruim:** não dá pra abrir no Notepad — só com ferramenta certa (pandas, DuckDB, Spark).

Esse é o formato que vamos usar quase sempre nas demos.

## Comparação de tamanhos

Mesmo dado, três formatos. Quem é menor?

```python
import os

for nome in ["clientes_pequeno.csv", "clientes_pequeno.json", "clientes_pequeno.parquet"]:
    caminho = f"data/{nome}"
    tamanho = os.path.getsize(caminho)
    print(f"{nome:<26} {tamanho:>6} bytes")
```

Com 100 linhas a diferença é pequena. Com **milhões** de linhas, Parquet pode ser **10x menor** que CSV.

## Quando usar cada um

| Formato | Use quando |
|---------|-----------|
| **CSV** | Compartilhar com não-técnicos · Importar em planilhas |
| **JSON** | Conversar com APIs · Dados com estrutura aninhada |
| **Parquet** | Análise séria · Volumes grandes · Performance importa |
